```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE
sns.set_theme(style='whitegrid')
```

## Prepare Features For Clustering


```python
work = df.copy()

# basic prep
work['post_date'] = pd.to_datetime(work.get('post_date'), errors='coerce')
if 'posting_hour' not in work.columns:
    work['posting_hour'] = work['post_date'].dt.hour
if 'day_of_week' not in work.columns:
    work['day_of_week'] = work['post_date'].dt.day_name()
if 'month' not in work.columns:
    work['month'] = work['post_date'].dt.month

for c in ['followers_count','likes_count','comments_count','views_count','posting_hour','month']:
    if c in work.columns:
        work[c] = pd.to_numeric(work[c], errors='coerce')

if 'engagement_rate' not in work.columns:
    work['engagement_rate'] = np.where(
        work['followers_count'] > 0,
        (work['likes_count'].fillna(0) + work['comments_count'].fillna(0)) / work['followers_count'],
        np.nan
    )
else:
    work['engagement_rate'] = pd.to_numeric(work['engagement_rate'], errors='coerce')

if 'view_rate' not in work.columns:
    work['view_rate'] = np.where(
        work['followers_count'] > 0,
        work['views_count'].fillna(0) / work['followers_count'],
        np.nan
    )
else:
    work['view_rate'] = pd.to_numeric(work['view_rate'], errors='coerce')

work['day_of_week'] = work['day_of_week'].astype(str)
work['sector'] = work['sector'].astype(str)
work['post_type'] = work['post_type'].astype(str)

# aggregate slots (keep posting_hour for best-time resolution)
slot_df = (
    work.groupby(['sector','post_type','day_of_week','posting_hour'], dropna=False)
    .agg(
        avg_engagement_rate=('engagement_rate','mean'),
        avg_view_rate=('view_rate','mean'),
        avg_likes=('likes_count','mean'),
        avg_comments=('comments_count','mean'),
        avg_views=('views_count','mean'),
        post_count=('engagement_rate','size')
    )
    .reset_index()
)

weekday_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
weekday_map = {d:i for i,d in enumerate(weekday_order)}
slot_df['day_idx'] = slot_df['day_of_week'].map(weekday_map)

num_cols = ['posting_hour','day_idx','avg_engagement_rate','avg_view_rate','post_count']
X = slot_df[num_cols].copy()

# robust NaN/Inf handling for JSON-based data
for c in num_cols:
    X[c] = pd.to_numeric(X[c], errors='coerce')
X = X.replace([np.inf, -np.inf], np.nan)

# fill each column safely (median if available, else 0)
for c in num_cols:
    med = X[c].median()
    if pd.isna(med):
        med = 0
    X[c] = X[c].fillna(med)

# final safety: if any NaN remain, fill with 0
X = X.fillna(0)

# sanity checks
print('Any NaN in X:', bool(X.isna().any().any()))
print('Any Inf in X:', bool(np.isinf(X.to_numpy()).any()))
print('Clustering rows:', len(slot_df))
print('Features used for clustering:', num_cols)

X_scaled = StandardScaler().fit_transform(X)

```

## KMeans (Elbow Method)


```python
# KMeans (tuned) on original scaled features
inertias = []
sils = []
K = range(2, min(10, len(X_scaled)))
for k in K:
    km = KMeans(n_clusters=k, random_state=42, n_init=30)
    lbl = km.fit_predict(X_scaled)
    inertias.append(km.inertia_)
    sils.append(silhouette_score(X_scaled, lbl))
fig, ax = plt.subplots(1,2, figsize=(12,4))
ax[0].plot(list(K), inertias, marker='o')
ax[0].set_title('Elbow Method (KMeans)')
ax[0].set_xlabel('k')
ax[0].set_ylabel('Inertia')
ax[0].grid(True, alpha=0.3)
ax[1].plot(list(K), sils, marker='o', color='darkgreen')
ax[1].set_title('Silhouette by k')
ax[1].set_xlabel('k')
ax[1].set_ylabel('Silhouette')
ax[1].grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
# choose best k by silhouette (more stable than fixed k)
k_final = list(K)[int(np.argmax(sils))]
kmeans = KMeans(n_clusters=k_final, random_state=42, n_init=30)
slot_df['kmeans_cluster'] = kmeans.fit_predict(X_scaled)
print('KMeans k used (best silhouette):', k_final)
print(slot_df['kmeans_cluster'].value_counts().sort_index())
```

## DBSCAN


```python
# DBSCAN on t-SNE space (matches visual cluster structure)
perplexity = max(5, min(30, len(X_scaled)//10))
tsne_db = TSNE(n_components=2, random_state=42, perplexity=perplexity, learning_rate='auto', init='pca')
X2_db = tsne_db.fit_transform(X_scaled)
rows = []
# tighter grid for 2D embedding
for eps in np.round(np.linspace(0.7, 3.0, 50), 3):
    for ms in [3,4,5,6,8]:
        db = DBSCAN(eps=float(eps), min_samples=int(ms))
        lbl = db.fit_predict(X2_db)
        n_clusters = len(set(lbl) - {-1})
        n_noise = int((lbl == -1).sum())
        noise_ratio = n_noise / len(lbl)
        sil = np.nan
        if n_clusters >= 2:
            mask = lbl != -1
            if mask.sum() > n_clusters and len(np.unique(lbl[mask])) > 1:
                sil = silhouette_score(X2_db[mask], lbl[mask])
        # prefer 3..5 visual groups with low noise
        penalty = 0.0
        if n_clusters < 3:
            penalty = (3 - n_clusters) * 0.25
        elif n_clusters > 5:
            penalty = (n_clusters - 5) * 0.20
        score = np.nan
        if not np.isnan(sil):
            score = sil - 0.40*noise_ratio - penalty
        rows.append({
            'eps': float(eps),
            'min_samples': int(ms),
            'n_clusters': int(n_clusters),
            'noise': int(n_noise),
            'noise_ratio': float(noise_ratio),
            'sil': sil,
            'score': score,
            'labels': lbl
        })
search = pd.DataFrame(rows)
cand = search[(search['n_clusters'] >= 2) & (~search['sil'].isna())].copy()
if len(cand) == 0:
    best = search.sort_values(['noise', 'n_clusters'], ascending=[True, False]).iloc[0].to_dict()
    mode = 'fallback'
else:
    pref = cand[(cand['n_clusters'].between(3,5)) & (cand['noise_ratio'] <= 0.18)].copy()
    if len(pref) > 0:
        best = pref.sort_values(['sil','noise_ratio','score'], ascending=[False,True,False]).iloc[0].to_dict()
        mode = 'preferred_3to5_low_noise'
    else:
        best = cand.sort_values(['score','noise_ratio','sil'], ascending=[False,True,False]).iloc[0].to_dict()
        mode = 'best_available'
slot_df['dbscan_cluster'] = best['labels']
print('DBSCAN mode:', mode)
print('Best DBSCAN:', {
    'eps': best['eps'],
    'min_samples': best['min_samples'],
    'n_clusters': best['n_clusters'],
    'noise': best['noise'],
    'noise_ratio': round(best['noise_ratio'], 3),
    'sil': None if np.isnan(best['sil']) else round(float(best['sil']), 4)
})
print(slot_df['dbscan_cluster'].value_counts().sort_index())
```

## Compare KMeans vs DBSCAN + Visualization


```python
# comparison numbers
km_sil = silhouette_score(X_scaled, slot_df['kmeans_cluster']) if len(np.unique(slot_df['kmeans_cluster'])) > 1 else np.nan
db_labels = slot_df['dbscan_cluster'].to_numpy()
db_non_noise = db_labels != -1
if len(set(db_labels[db_non_noise])) > 1 and db_non_noise.sum() > len(set(db_labels[db_non_noise])):
    # evaluate DBSCAN on same t-SNE space used for fitting
    db_sil = silhouette_score(X2_db[db_non_noise], db_labels[db_non_noise])
else:
    db_sil = np.nan
print({'kmeans_sil': km_sil, 'dbscan_sil_non_noise': db_sil})
# 2D view: use same t-SNE used by DBSCAN
vis = pd.DataFrame({'x': X2_db[:,0], 'y': X2_db[:,1], 'kmeans': slot_df['kmeans_cluster'], 'dbscan': slot_df['dbscan_cluster']})
fig, ax = plt.subplots(1,2, figsize=(13,5), sharex=True, sharey=True)
sns.scatterplot(data=vis, x='x', y='y', hue='kmeans', palette='tab10', ax=ax[0], s=45)
ax[0].set_title(f'KMeans Clusters (k={slot_df['kmeans_cluster'].nunique()})')
sns.scatterplot(data=vis, x='x', y='y', hue='dbscan', palette='tab10', ax=ax[1], s=45)
ax[1].set_title('DBSCAN Clusters')
plt.tight_layout(); plt.show()
```
